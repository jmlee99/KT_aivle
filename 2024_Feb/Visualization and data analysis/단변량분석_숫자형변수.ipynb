### **(1) 수치화 : 대푯값**
1. 평균(mean)
2. 중앙값(median)
3. 최빈값(mode)
4. 4분위수(quantile)

```python
# 1. 평균(산술평균)
# 넘파이 함수 이용하기
np.mean(titanic['Fare'])
```

```python
# 판다스의 mean 메서드 이용하기
titanic['Fare'].mean()
```

```python
# 2. 중앙값(중위수, median)
# 자료의 순서상 가운데 위치한 값
np.median(titanic['Fare'])
```

```python
# 판다스의 median 메서드 이용하기
titanic['Fare'].median()
```

```python
# 3. 최빈값
# 자료 중에서 가장 빈번하게 나타나는 값(빈도가 높은 값)
# 이산형 데이터(셀 수 있는 데이터)
# 판다스의 mode 메서드 이용하기
titanic['Pclass'].mode()
```

### **(3) 시각화**

#### **1) 히스토그램**

```python
plt.hist(titanic.Fare, bins = 5, edgecolor = 'gray')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# 다른 방법
fig, axe = plt.subplots()
axe.hist(titanic['Fare'], bins = 5, edgecolor = 'blue')
axe.set_xlabel('Fare')
axe.set_ylabel('Frequency')
```

```python
# 구간의 갯수를 조정해 보자
plt.hist(titanic.Fare, bins = 30, edgecolor = 'gray')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
#구간의 갯수 조절 다른 방법
fig, axe = plt.subplots()
axe.hist(titanic.Fare, bins = 30, edgecolor = 'blue')
axe.set_title('Fare')
axe.set_title('Frequency')
```

### **sns.histplot**

```python
sns.histplot(data = titanic, x = 'Fare', bins = 20)
```

* (추가)구간 값과 갯수를 확인해 봅시다.

    * 결과를 저장하면, 구간과 빈도수를 확인할 수 있습니다.
        * 히스토그램을 저장하면 튜플.
        * 튜플 안에는 넘파이 어레이가 2개로 구성됨.
        * 첫번째 넘파이 어레이는 빈도수,
        * 두번째는 구간 값.


```python
hist1 = plt.hist(titanic.Fare, bins = 5, edgecolor = 'gray')
plt.show()

print(hist1)
print(type(hist1))
print('-' * 50)
print('빈도수 : ', hist1[0])
print('구간값 : ', hist1[1])
```

#### **2) 밀도함수 그래프(kde plot)**

* 히스토그램의 단점
    * 구간(bin)의 너비를 어떻게 잡는지에 따라 전혀 다른 모양이 될 수 있음
>
* 밀도함수 그래프
    * 막대의 너비를 가정하지 않고 모든 점에서 데이터의 밀도를 추정하는 커널 밀도 추정(Kernel Density Estimation)방식을 사용하여 이러한 단점을 해결.
    * 밀도함수 그래프 아래 면적은 1

```python
sns.kdeplot(titanic['Fare'])
# sns.kdeplot(x='Fare', data = titanic)

plt.show()
```

- 밀도 추정

    * 측정된 데이터로부터 전체 데이터 분포의 특성을 추정

<img src='https://github.com/DA4BAM/image/blob/main/%EB%B0%80%EB%8F%84%EC%B6%94%EC%A0%95.png?raw=true' width=800/>

```python
# 문제1 titanic Age에 대해서 밀도함수 그래프를 그려봅시다.
# 히스토그램과 어떤 차이가 있나요?

sns.histplot(data = titanic, x = 'Age', bins = 20)
```

```python
# 문제2 titanic Age에 대해서 히스토그램을 그려서
# 단, 이번에는 sns.histplot
# 옵션은 kde = True를 지정해 보자

sns.histplot(data = titanic, x = 'Age' , bins = 20, kde = True)
```

#### **3) boxplot**

**주의사항** : 값에 **NaN**이 있으면 그래프가 그려지지 않습니다.

- boxplot 기본

```python
# titanic['Age']에는 NaN이 있습니다. 이를 제외한 데이터
temp = titanic.loc[titanic['Age'].notnull()]
```

```python
plt.boxplot(temp['Age'])
plt.grid()
plt.show()
```

```python
# 다른 방법
fig, axe = plt.subplots()
axe.boxplot(temp['Age'])
plt.grid()
```

- 옆으로 그리기

```python
plt.boxplot(temp['Age'], vert = False)
plt.grid()
plt.show()
```

```python
# 다른 방법 2

fig, axe = plt.subplots()
axe.boxplot(temp['Age'], vert = False)
plt.grid()
```

- ***sns.boxplot***
- ***seaborn 패키지 함수들은 NaN을 알아서 뺴준다.***

```python
sns.boxplot(x = titanic['Age'])
plt.grid()
plt.show()
```

```python
sns.boxplot(y = titanic['Age'])
plt.grid()
plt.show()
```

```python
# 문제1 titanic의 Age와 Fare에 대해서 각각 박스 플롯을 그리고 해석해보자

fig, axes = plt.subplots(1,2, figsize = (12,8))
sns.boxplot(data = titanic, y = 'Fare', ax = axes[0])
sns.boxplot(data = titanic, y = 'Age', ax = axes[1])

axes[0].set_title('Fare')
axes[1].set_title('Age')
```

```python
plt.figure(figsize = (8,6)) 
plt.subplot(2,1,1) 
sns.histplot(x= 'Fare', data = titanic, bins = 30, kde = True) 
plt.grid() 
plt.subplot(2,1,2) 
sns.boxplot(x= 'Fare', data = titanic) 
plt.grid() 
plt.tight_layout() 
plt.show()
```
