## **3.단변량분석 : 숫자형 변수**

```python
def eda_1_num(data, var, bins = 30):

    # 기초통계량
    print('<< 기초통계량 >>')
    display(data[[var]].describe().T)
    print('=' * 100)

    # 시각화
    print('<< 그래프 >>')
    plt.figure(figsize = (10,6))

    plt.subplot(2,1,1)
    sns.histplot(data[var], bins = bins, kde = True)
    plt.grid()

    plt.subplot(2,1,2)
    sns.boxplot(x = data[var])
    plt.grid()
    plt.show()
```

### **(1) Sales (Target)**

```python
var = 'Sales'
```

```python
eda_1_num(data, var)
```

## **4.단변량분석 : 범주형 변수**

### **(1) ShelveLoc** 

```python
def eda_1_cat(data, var, str1) :
    t1 = data[var].value_counts()
    t2 = data[var].value_counts(normalize = True)
    t3 = pd.concat([t1, t2], axis = 1)
    t3.columns = ['count','ratio']
    display(t3)
    sns.countplot(x = var, data = data)
    plt.title(str1)
    plt.show()
    
```

```python
var = 'ShelveLoc'
str1 = '매장 진열상태'
```

```python
eda_1_cat(data, var, str1)
```

### **(2) Urban**

```python
var = 'Urban'
str1 = '도심 지역 여부(Yes, No)'
```

```python
eda_1_cat(data, var, str1)
```

### **(3) US**

```python
var = 'US'
str1 = '(미국) 매장 유무'
```

```python
eda_1_cat(data, var, str1)
```
