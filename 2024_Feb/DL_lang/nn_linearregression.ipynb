```python
import torch
x_train = torch.FloatTensor([[1,1], [2,2], [3,3]])
y_train = torch.FloatTensor([[10], [20], [30]])
W = torch.randn([2,1], requires_grad = True)
b = torch.randn([1], requires_grad = True)
optimizer = torch.optim.SGD([W,b], lr = 0.01)
# [딥러닝의 1단계] : 모델을 만든다.
def H(x):
    return torch.matmul(x, W) + b # H(x) = Wx + b
# [딥러닝의 2단계] : 학습을 한다.
for step in range(2000):
    cost = torch.mean((H(x_train)-y_train) ** 2) # cost - 평균((H(x)-y의 제곱)
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
# [딥러닝의 3단계] : 추론/테스트를 한다.
x_test = torch.FloatTensor([4,4])
print(H(x_test))
print(H(x_test).detach())
print(H(x_test).detach().item())
```
